```python
E0=x0

Ej=mb*B*(2.+2*l)/(1+2*l)*(l+1/2-k)*x
Ejn=mb*B*(2.*l)/(1+2*l)*(l-1/2-k)*x

Ej1=mb*B*(2.+2*l1)/(1+2*l1)*(l1+1/2-k1)*x
Ejn1=mb*B*(2.*l1)/(1+2*l1)*(l1-1/2-k1)*x

Ej0=mb*B*(2.+2*l0)/(1+2*l0)*(l0+1/2-k0)*x
Ejn0=mb*B*(2.*l0)/(1+2*l0)*(l0-1/2-k0)*x


E0B=x0

EjB=mb*B2*(2.+2*l)/(1+2*l)*(l+1/2-k)*x
EjnB=mb*B2*(2.*l)/(1+2*l)*(l-1/2-k)*x


Ej1B=mb*B2*(2.+2*l1)/(1+2*l1)*(l1+1/2-k1)*x
Ejn1B=mb*B2*(2.*l1)/(1+2*l1)*(l1-1/2-k1)*x

Ej0B=mb*B2*(2.+2*l0)/(1+2*l0)*(l0+1/2-k0)*x
Ejn0B=mb*B2*(2.*l0)/(1+2*l0)*(l0-1/2-k0)*x

print(Ej)
print(Ej1)
print(Ej0)
```
