```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import math
%matplotlib inline
```

```python
#These Plots are all for Hydrogen

#At the moment:
mb=1
B=1
B2=4
x0=0
x=1
y=5

```

```python
l=2
l1=1
l0=0
```

```python
k=np.arange(0,2*l+1,1)
kn=np.arange(0,2*l-1,1)

k1=np.arange(0,2*l1+1,1)
kn1=np.arange(0,2*l1-1,1)

k0=np.arange(0,2*l0+1,1)
kn0=np.arange(0,2*l0-1,1)
```

```python
E0=x0

Ej=mb*B*(2.+2*l)/(1+2*l)*(l+1/2-k)*x
Ejn=mb*B*(2.*l)/(1+2*l)*(l-1/2-k)*x

Ej1=mb*B*(2.+2*l1)/(1+2*l1)*(l1+1/2-k1)*x
Ejn1=mb*B*(2.*l1)/(1+2*l1)*(l1-1/2-k1)*x

Ej0=mb*B*(2.+2*l0)/(1+2*l0)*(l0+1/2-k0)*x
Ejn0=mb*B*(2.*l0)/(1+2*l0)*(l0-1/2-k0)*x


E0B=x0

EjB=mb*B2*(2.+2*l)/(1+2*l)*(l+1/2-k)*x
EjnB=mb*B2*(2.*l)/(1+2*l)*(l-1/2-k)*x

Ej1B=mb*B2*(2.+2*l1)/(1+2*l1)*(l1+1/2-k1)*x
Ejn1B=mb*B2*(2.*l1)/(1+2*l1)*(l1-1/2-k1)*x

Ej0B=mb*B2*(2.+2*l0)/(1+2*l0)*(l0+1/2-k0)*x
Ejn0B=mb*B2*(2.*l0)/(1+2*l0)*(l0-1/2-k0)*x

print(Ej)
print(Ej1)
print(Ej0)
```

```python
fig1 = plt.figure(figsize=(10,10))
#Energy Before Splitting
plt.plot(x0,E0) #j=l-1/2 state initial energy
plt.plot(x0,E0+y) #j=l+1/2 state initial energy

#j=l+1/2 State Splitting
#l=2
plt.plot(x,Ej[0]+y,'ko',label="l=2")
plt.plot(x,Ej[1]+y,'ko')
plt.plot(x,Ej[2]+y,'ko')
plt.plot(x,Ej[3]+y,'ko')
plt.plot(x,Ej[4]+y,'ko')
#l=1
plt.plot(x,Ej1[0]+y,'ro',label="l=1")
plt.plot(x,Ej1[1]+y,'ro')
plt.plot(x,Ej1[2]+y,'ro')
#l=0
plt.plot(x,Ej0[0]+y,'yo',label="l=0")



#Reference Lines
plt.plot([0.3, 0.6], [y, y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej[0]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej[1]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej[2]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej[3]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej[4]+y], color='k', linestyle='-', linewidth=2)

plt.text(0.3, y+0.2, 'j=l+1/2', fontsize=30)

plt.text(0.3, 0.2, 'j=l-1/2', fontsize=30)

plt.text(0.55, y+3, 'B=1', fontsize=30)

plt.plot([0.6, 1], [y, Ej1[0]+y], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej1[1]+y], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej1[2]+y], color='r', linestyle='-', linewidth=2)

plt.plot([0.6, 1], [y, Ej0[0]+y], color='y', linestyle='-', linewidth=2)

plt.plot([0.3, 0.6], [0, 0], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn[0]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn[1]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn[2]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn[3]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn[4]], color='k', linestyle='-', linewidth=2)

plt.plot([0.6, 1], [0, Ejn1[0]], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn1[1]], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn1[2]], color='r', linestyle='-', linewidth=2)

plt.plot([0.6, 1], [0, Ejn0[0]], color='y', linestyle='-', linewidth=2)


#j=l-1/2 State Splitting
#l=2
plt.plot(x,Ejn[0],'ko')
plt.plot(x,Ejn[1],'ko')
plt.plot(x,Ejn[2],'ko')
plt.plot(x,Ejn[3],'ko')
plt.plot(x,Ejn[4],'ko')
#l=1
plt.plot(x,Ejn1[0],'ro')
plt.plot(x,Ejn1[1],'ro')
plt.plot(x,Ejn1[2],'ro')
#l=0
plt.plot(x,Ejn0[0],'yo')


plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.axis([0.25,1.05, -2, 10],fontsize=30)
plt.ylabel('Energy Levels',fontsize=30)
plt.legend(fontsize=30)










fig2 = plt.figure(figsize=(10,10))
#Exact Same Thing but with another B Field.
#Energy Before Splitting
plt.plot(x0,E0B) #j=l-1/2 state initial energy
plt.plot(x0,E0B+y) #j=l+1/2 state initial energy

#j=l+1/2 State Splitting
#l=2
plt.plot(x,EjB[0]+y,'ko',label="l=2")
plt.plot(x,EjB[1]+y,'ko')
plt.plot(x,EjB[2]+y,'ko')
plt.plot(x,EjB[3]+y,'ko')
plt.plot(x,EjB[4]+y,'ko')
#l=1
plt.plot(x,Ej1B[0]+y,'ro',label="l=1")
plt.plot(x,Ej1B[1]+y,'ro')
plt.plot(x,Ej1B[2]+y,'ro')
#l=0
plt.plot(x,Ej0B[0]+y,'yo',label="l=0")



#Reference Lines
plt.plot([0.3, 0.6], [y, y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, EjB[0]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, EjB[1]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, EjB[2]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, EjB[3]+y], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, EjB[4]+y], color='k', linestyle='-', linewidth=2)

plt.text(0.3, y+0.2, 'j=l+1/2', fontsize=30)

plt.text(0.3, 0.2, 'j=l-1/2', fontsize=30)

plt.text(0.55, y+5, 'B=4', fontsize=30)

plt.plot([0.6, 1], [y, Ej1B[0]+y], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej1B[1]+y], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [y, Ej1B[2]+y], color='r', linestyle='-', linewidth=2)

plt.plot([0.6, 1], [y, Ej0B[0]+y], color='y', linestyle='-', linewidth=2)

plt.plot([0.3, 0.6], [0, 0], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, EjnB[0]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, EjnB[1]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, EjnB[2]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, EjnB[3]], color='k', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, EjnB[4]], color='k', linestyle='-', linewidth=2)

plt.plot([0.6, 1], [0, Ejn1B[0]], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn1B[1]], color='r', linestyle='-', linewidth=2)
plt.plot([0.6, 1], [0, Ejn1B[2]], color='r', linestyle='-', linewidth=2)

plt.plot([0.6, 1], [0, Ejn0B[0]], color='y', linestyle='-', linewidth=2)


#j=l-1/2 State Splitting
#l=2
plt.plot(x,EjnB[0],'ko')
plt.plot(x,EjnB[1],'ko')
plt.plot(x,EjnB[2],'ko')
plt.plot(x,EjnB[3],'ko')
plt.plot(x,EjnB[4],'ko')
#l=1
plt.plot(x,Ejn1B[0],'ro')
plt.plot(x,Ejn1B[1],'ro')
plt.plot(x,Ejn1B[2],'ro')
#l=0
plt.plot(x,Ejn0B[0],'yo')

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.axis([0.25,1.05, -8, 15],fontsize=30)
plt.ylabel('Energy Levels',fontsize=30)
plt.legend(fontsize=30)
```
